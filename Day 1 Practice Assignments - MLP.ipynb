# Implemet a multilayer perceptron - practice Assignments

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Define the model
def create_mlp(input_size, hidden_size, output_size):
    model = Sequential()
    model.add(Dense(hidden_size, input_dim=input_size, activation='sigmoid'))
    model.add(Dense(output_size, activation='sigmoid'))
    return model

# Example usage
input_size = 2
hidden_size = 3
output_size = 1

mlp_model = create_mlp(input_size, hidden_size, output_size)

# Display the model architecture
mlp_model.summary()

# Compile the model
mlp_model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])

# Training data
inputs = tf.constant([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=tf.float32)
targets = tf.constant([[0], [1], [1], [0]], dtype=tf.float32)

# Training the model
mlp_model.fit(inputs, targets, epochs=1000, verbose=2)

# Testing the trained model
predictions = mlp_model.predict(inputs)
for i in range(len(inputs)):
    print(f"Input: {inputs[i].numpy()}, Predicted Output: {predictions[i][0]}")
```
